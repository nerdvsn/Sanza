## Plancksches Strahlungsgesetz (spektrale Strahldichte)

Das Plancksche Strahlungsgesetz beschreibt die spektrale Verteilung der Strahlung eines idealen schwarzen Körpers in Abhängigkeit von der Wellenlänge \( \lambda \) und der Temperatur \( T \):

$$
B(\lambda, T) = \frac{2 h c^2}{\lambda^5} \cdot \frac{1}{\exp\left( \frac{h c}{\lambda k_B T} \right) - 1}
$$

---

## Bedeutung der Variablen


\begin{array}{|c|l|l|}
\hline
\textbf{Symbol} & \textbf{Bedeutung} & \textbf{Einheit} \\
\hline
B(\lambda, T) & \text{Strahldichte (Intensität pro Wellenlänge)} & \mathrm{W \cdot m^{-3} \cdot sr^{-1}} \\
\lambda & \text{Wellenlänge} & \text{Meter (m)} \\
T & \text{Temperatur des Körpers} & \text{Kelvin (K)} \\
h & \text{Plancksches Wirkungsquantum} & 6{,}626 \times 10^{-34} \ \mathrm{J \cdot s} \\
c & \text{Lichtgeschwindigkeit} & 2{,}998 \times 10^8 \ \mathrm{m/s} \\
k_B & \text{Boltzmann-Konstante} & 1{,}381 \times 10^{-23} \ \mathrm{J/K} \\
\hline
\end{array}



## Wiensches Verschiebungsgesetz

Das Wiensche Verschiebungsgesetz beschreibt den Zusammenhang zwischen der Temperatur eines schwarzen Körpers und der Wellenlänge, bei der dieser die maximale Strahlungsintensität aussendet.

**Aussage:**

Je heißer ein Körper, desto kürzer ist die Wellenlänge, bei der er am meisten strahlt.

**Formel:**

$$
\lambda_{\text{max}} = \frac{b}{T}
$$

---

## Bedeutung der Variablen

\begin{array}{|c|l|l|}
\hline
\textbf{Symbol} & \textbf{Bedeutung} & \textbf{Einheit} \\
\hline
\lambda_{\text{max}} & \text{Wellenlänge mit maximaler Intensität} & \mathrm{m} \\
T & \text{Temperatur des Körpers} & \mathrm{K} \\
b & \text{Wiensche Verschiebungskonstante} & 2{,}897 \times 10^{-3} \ \mathrm{m \cdot K} \\
\hline
\end{array}


---

**Merksatz:**

Ein heißer Körper sendet mehr Strahlung aus – und tut das bei kürzeren Wellenlängen.


## Warum bedeuten kürzere Wellenlängen mehr Energie?

Bei elektromagnetischer Strahlung (z. B. Licht) gilt folgender Zusammenhang zwischen der Energie eines Photons und seiner Wellenlänge:

$$
E = \frac{h c}{\lambda}
$$

---

## Bedeutung der Variablen


\begin{array}{|c|l|l|}
\hline
\textbf{Symbol} & \textbf{Bedeutung} & \textbf{Einheit} \\
\hline
E & \text{Energie eines Photons} & \mathrm{J} \ (\text{Joule}) \\
h & \text{Plancksches Wirkungsquantum} & 6{,}626 \times 10^{-34} \ \mathrm{J \cdot s} \\
c & \text{Lichtgeschwindigkeit} & 2{,}998 \times 10^8 \ \mathrm{m/s} \\
\lambda & \text{Wellenlänge} & \mathrm{m} \\
\hline
\end{array}


---

**Fazit:**

Wenn \( \lambda \) kleiner wird, dann wird \( E \) größer.

**Folglich gilt:**  
**Kürzere Wellenlänge = mehr Energie**


## Stefan-Boltzmann-Gesetz

Das Stefan-Boltzmann-Gesetz beschreibt die gesamte Energieabstrahlung eines Schwarzen Körpers in Abhängigkeit von seiner Temperatur.

**Aussage:**

Die gesamte abgestrahlte Energie eines Schwarzen Körpers pro Fläche steigt mit der vierten Potenz seiner Temperatur.

---

## Mathematische Form

**Gesamte Strahlungsleistung:**

$$
P = \sigma \cdot A \cdot T^4
$$

**Strahlungsleistung pro Fläche:**

$$
j^* = \sigma \cdot T^4
$$

---

## Bedeutung der Variablen

\[
\begin{array}{|c|l|l|}
\hline
\textbf{Symbol} & \textbf{Bedeutung} & \textbf{Einheit} \\
\hline
P & \text{Gesamte Strahlungsleistung} & \mathrm{W} \ (\text{Watt}) \\
A & \text{Oberfläche des Körpers} & \mathrm{m^2} \\
T & \text{Temperatur des Körpers} & \mathrm{K} \ (\text{Kelvin}) \\
j^* & \text{Abgestrahlte Leistung pro Fläche} & \mathrm{W/m^2} \\
\sigma & \text{Stefan{-}Boltzmann{-}Konstante} & 5{,}670 \times 10^{-8} \ \mathrm{W/m^2 \cdot K^4} \\
\hline
\end{array}
\]

---

## Bedeutung in Worten

Wenn du die Temperatur verdoppelst, wird die abgestrahlte Leistung **16-mal so groß**, da:

$$
(2T)^4 = 16 \cdot T^4
$$

**Heißere Objekte strahlen extrem viel mehr Energie ab.**

---

## Beispiel

Ein Objekt mit:

- Temperatur: \( T = 300 \, \mathrm{K} \) (z. B. ein Mensch)  
- Fläche: \( A = 1 \, \mathrm{m^2} \)

Berechnung der Strahlungsleistung:

$$
P = \sigma \cdot A \cdot T^4 = 5{,}67 \times 10^{-8} \cdot 1 \cdot (300)^4 \approx 459 \, \mathrm{W}
$$

**Ergebnis:**  
Dein Körper gibt rund **459 Watt** an Wärmestrahlung ab – wie ein kleiner Radiator!

---

## Zusammenhang mit dem Planckschen Gesetz

Das Plancksche Gesetz beschreibt die **Strahlung pro Wellenlänge**.

Das Stefan-Boltzmann-Gesetz ergibt sich durch **Integration über alle Wellenlängen**:

$$
j^* = \int_0^\infty B(\lambda, T) \, d\lambda = \sigma T^4
$$

---

## Merksatz

**Je heißer ein Körper, desto viel mehr Strahlung – mit \( T^4 \)!**


## Was ist das Plancksche Wirkungsquantum \( h \) ?

Das Plancksche Wirkungsquantum ist eine fundamentale Naturkonstante.  
Es beschreibt den kleinsten möglichen Energieaustausch zwischen Licht (Photonen) und Materie – also: **Energie ist quantisiert**.

---

## Mathematische Definition

Der Zusammenhang zwischen Energie und Frequenz ist:

$$
E = h \cdot f
$$

---

## Bedeutung der Variablen

\[
\begin{array}{|c|l|l|}
\hline
\textbf{Symbol} & \textbf{Bedeutung} & \textbf{Einheit} \\
\hline
E & \text{Energie eines Photons (Lichtteilchens)} & \mathrm{J} \ (\text{Joule}) \\
h & \text{Plancksches Wirkungsquantum} & 6{,}626 \times 10^{-34} \ \mathrm{J \cdot s} \\
f & \text{Frequenz der Strahlung} & \mathrm{Hz} \ (\text{1/s}) \\
\hline
\end{array}
\]

---

## Warum ist das wichtig?

Ohne \( h \) hätte man nicht verstanden, warum glühende Körper bei kurzen Wellenlängen **nicht unendlich viel Energie abstrahlen** – das sogenannte **Ultraviolett-Katastrophen-Problem** der klassischen Physik.

Planck führte \( h \) ein, um die Schwarzkörperstrahlung korrekt zu beschreiben.  
Einstein verwendete es später zur Erklärung des **photoelektrischen Effekts** – wofür er den **Nobelpreis** erhielt.

---

## Veranschaulichung – Analogie

Stell dir ein Klavier vor:  
Du kannst **nicht jede beliebige Tonhöhe** spielen – nur **bestimmte diskrete Töne**.  
So ist es auch mit Energie auf quantenphysikalischer Ebene:  
**Energie kommt nicht kontinuierlich, sondern in diskreten Paketen.**

Die Größe dieser Energiepakete hängt von \( h \) ab.

---

## Beispiel: Licht und Photonen

- Ein **blaues Photon** (hohe Frequenz \( f \)) hat **mehr Energie**  
- Ein **rotes Photon** (niedrige Frequenz \( f \)) hat **weniger Energie**

Die Energie ist jeweils:

$$
E = h \cdot f
$$

---

## Verbindung zur Planckschen Strahlungsgleichung

In der Planckschen Formel steckt \( h \) direkt drin:

$$
B(\lambda, T) = \frac{2 h c^2}{\lambda^5} \cdot \frac{1}{\exp\left( \frac{h c}{\lambda k_B T} \right) - 1}
$$

→ Ohne \( h \) gäbe es **keine sinnvolle Beschreibung thermischer Strahlung**!

---

## Merksatz

Das Plancksche Wirkungsquantum ist die **kleinste Portion Energie**, die Licht tragen kann.

In der Quantenwelt ist Energie **nicht fließend**, sondern **häppchenweise** – wie Treppenstufen.


## Strahlungstransportgleichung

Die Strahlungstransportgleichung beschreibt, welche Strahlung bei einer bestimmten Wellenlänge \( \lambda \) tatsächlich von einem Objekt gemessen wird – als Kombination aus **emittierter Eigenstrahlung** und **reflektierter Umgebungsstrahlung**.

### Mathematische Form:

$$
S(\lambda) = \varepsilon(\lambda) \cdot B(\lambda, T) + \left(1 - \varepsilon(\lambda)\right) \cdot R(\lambda)
$$

---

## Bedeutung der Variablen

\[
\begin{array}{|c|l|}
\hline
\textbf{Symbol} & \textbf{Bedeutung} \\
\hline
S(\lambda) & \text{Gemessene Strahlung bei Wellenlänge } \lambda \\
\varepsilon(\lambda) & \text{Emissivität des Objekts bei Wellenlänge } \lambda \ (0 \leq \varepsilon \leq 1) \\
B(\lambda, T) & \text{Plancksche Strahlung bei Temperatur } T \\
R(\lambda) & \text{Reflektierte Umgebungsstrahlung bei } \lambda \ (\text{z. B. Sonnenlicht, Wände}) \\
1 - \varepsilon(\lambda) & \text{Anteil der Strahlung, der nicht emittiert, sondern reflektiert wird} \\
\hline
\end{array}
\]

---

## Was macht die Gleichung?

Die Gleichung teilt die gemessene Strahlung in zwei Hauptanteile auf:

### 1. **Eigenstrahlung (Emission)**

Das Objekt emittiert thermische Strahlung entsprechend seiner Temperatur:

$$
\varepsilon(\lambda) \cdot B(\lambda, T)
$$

### 2. **Reflexion der Umgebungsstrahlung**

Das Objekt reflektiert einen Teil der Strahlung aus der Umgebung:

$$
(1 - \varepsilon(\lambda)) \cdot R(\lambda)
$$

---

## Beispiel: Heiße Donut-Box

Stell dir eine Donut-Box aus glänzendem Metall vor:

- Sie ist warm (z. B. 50 °C)
- Die Sonne scheint darauf

Eine Wärmebildkamera misst nicht nur die Eigenwärme der Box, sondern auch Sonnenlicht, das reflektiert wird.

---

### Zerlegung der Messung:

\[
\begin{array}{|c|l|l|}
\hline
\textbf{Anteil} & \textbf{Quelle} & \textbf{Was du siehst} \\
\hline
\varepsilon(\lambda) \cdot B(\lambda, T) & \text{Wärme der Box} & \text{Eigene Strahlung (emittiert)} \\
(1 - \varepsilon(\lambda)) \cdot R(\lambda) & \text{Sonne / Umgebung} & \text{Reflektierte Strahlung (Umweltlicht)} \\
\hline
\end{array}
\]

---

## Warum ist das wichtig (z. B. in SANZA)?

In Anwendungen wie **SANZA** willst du die **wahre Temperatur \( T \)** eines Objekts ermitteln.  
Aber die Kamera sieht nur die gemischte Strahlung \( S(\lambda) \).  
→ Um daraus auf \( T \) zu schließen, brauchst du:

- die **Emissivität** \( \varepsilon(\lambda) \)
- die **reflektierte Umgebungsstrahlung** \( R(\lambda) \)

---

## Merksatz

**Was ich sehe = Was das Objekt selbst abstrahlt + Was es von außen reflektiert**
* Emissivität sagt, wie gut ein realer Körper Wärme abstrahlt – im Vergleich zu einem perfekten Schwarzen Körper.
