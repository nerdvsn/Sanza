## Erläuterung der Schritte im Code

---

### 1. Berechnung der Eigenstrahlung \( B(T) \)

Aus der gemessenen Strahlung \( S \), der Emissivität \( \varepsilon \) und der reflektierten Umgebungsstrahlung \( R \) wird die reine Eigenstrahlung des Objekts berechnet:

$$
B = \frac{S - (1 - \varepsilon) \cdot R}{\varepsilon}
$$

---

### 2. Temperaturberechnung über das Stefan–Boltzmann-Gesetz

Für graue Körper gilt der Zusammenhang zwischen Strahlung und Temperatur:

$$
B = \sigma \cdot T^4
\quad \Rightarrow \quad
T = \left( \frac{B}{\sigma} \right)^{\frac{1}{4}}
$$

---

### 3. Umrechnung von Kelvin in Grad Celsius

Die berechnete Temperatur \( T \) in Kelvin wird in Grad Celsius umgerechnet:

$$
T_{\text{[°C]}} = T_{\text{[K]}} - 273.15
$$


```python
# Beispiel 2: Strahlungstransportgleichung + Stefan–Boltzmann

import math

# Gegebene Größen
S = 4.50          # gemessene Strahlung S in W/m²
epsilon = 0.85    # Emissivität
R = 1.00          # reflektierte Strahlung R in W/m²

# Konstanten
sigma = 5.670e-8  # Stefan–Boltzmann-Konstante in W/m²·K⁴

# 1. Eigenstrahlung B(T) berechnen:
#    B = (S - (1 - ε) * R) / ε
B = (S - (1 - epsilon) * R) / epsilon

# 2. Temperatur aus Stefan–Boltzmann-Gesetz:
#    B = σ T⁴  ⇒  T = (B / σ)^(1/4)
T_kelvin = (B / sigma) ** 0.25

# 3. Umrechnung in °C:
T_celsius = T_kelvin - 273.15

# Ausgabe
print(f"Eigenstrahlung B(T) = {B:.2f} W/m²")
print(f"Temperatur T = {T_kelvin:.1f} K")
print(f"Temperatur T = {T_celsius:.2f} °C")
```

## Wie bekommst du die Emissivität \( \varepsilon \) ?

Die Emissivität ist **materialabhängig** und beschreibt, wie gut ein Material thermische Strahlung abgibt (0 = ideal reflektierend, 1 = ideal schwarz).

### Beispielwerte:

\[
\begin{array}{|l|c|}
\hline
\textbf{Material} & \textbf{Emissivität } \varepsilon \\
\hline
\text{Menschliche Haut} & 0{,}98 \\
\text{Beton} & 0{,}94 \\
\text{Rostfreier Stahl (poliert)} & 0{,}10 \text{ bis } 0{,}30 \\
\text{Wasser} & 0{,}96 \\
\hline
\end{array}
\]

Moderne Infrarotkameras verfügen oft über eine **interne Materialtabelle**, oder man stellt den Emissionswert manuell ein.

## Wie bekommst du die reflektierte Strahlung \( R \) ?

### Methode 1: Einfacher Ansatz

Messe die Umgebungstemperatur (z. B. mit einem Thermometer) und berechne die reflektierte Strahlung mit dem Stefan–Boltzmann-Gesetz:

$$
R = \sigma \cdot T_{\text{Umgebung}}^4
$$

**Beispiel:**

Umgebungstemperatur = \( 20 \,^\circ \text{C} = 293{,}15 \, \text{K} \)

$$
R = 5{,}670 \times 10^{-8} \cdot (293{,}15)^4 \approx 418{,}9 \, \text{W/m}^2
$$

---

### Methode 2: Profi-Ansatz (Reflexionsplatte)

- Stelle eine **Reflexions-Referenzfläche** auf (z. B. poliertes Aluminium)
- Miss mit der Kamera die scheinbare Temperatur dieser Fläche
- Rechne mit bekannter Emissivität und gemessener Strahlung zurück auf \( R \)

Diese Methode erlaubt eine **realitätsnähere Bestimmung der Umgebungsreflexion** – insbesondere in komplexen Szenen mit mehreren Strahlungsquellen.
